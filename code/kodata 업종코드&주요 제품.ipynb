# 업종코드별 주요제품

```python
# 업종코드와 주요제품 결측치 제거
df.dropna(subset=['업종코드', '주요제품'], inplace=True)
df.reset_index(drop=True)
```

```python
# '업종코드' 컬럼에서 앞 2자리를 추출하여 '업종분류' 컬럼을 생성
df['업종분류코드'] = df['업종코드'].astype(str).str[:3]
df[['업종코드', '업종분류코드', '주요제품']]

def extract_code(x):
    x_str = str(x)
    
    # 581만 3자리로 추출
    if x_str.startswith('581'):
        return x_str[:3]
    return x_str[:2]

# '업종코드' 컬럼에서 코드를 추출하여 '업종분류코드' 컬럼을 생성
df['업종분류코드'] = df['업종코드'].apply(extract_code)
df[['업종코드', '업종분류코드', '주요제품', '사업자번호']]
```

```python
def classify_sector(code):
    if code in ['10', '11', '12']:
        return '음식료'
    elif code in ['13', '14', '15']:
        return '섬유의복'
    elif code in ['16', '17', '18']:
        return '목재종이'
    elif code in ['19', '20', '21', '22']:
        return '석유화학'
    elif code == '23':
        return '비금속'
    elif code == '24':
        return '철강'
    elif code in ['25', '29']:
        return '기계'
    elif code in ['26', '27', '28']:
        return '전기전자'
    elif code in ['30', '31']:
        return '운송장비'
    elif code in ['32', '33', '34']:
        return '기타'
    elif code in ['58', '59', '60', '61', '62', '63']:
        if code == '581':
            return '출판업'
        return '정보통신'
    else:
        return '기타비제조'

def extract_code(x):
    x_str = str(x)
    
    # 581만 3자리로 추출
    if x_str.startswith('581'):
        return x_str[:3]
    return x_str[:2]

# '업종분류코드' 컬럼에서 코드를 추출하고 그 값을 기반으로 '업종분류' 컬럼을 생성
df['중분류'] = df['업종분류코드'].apply(extract_code).apply(classify_sector)
df[['업종코드', '업종분류코드', '중분류', '주요제품', '사업자번호']]
```

# 산업내 주력상품 판매비중

```python
# 1. 각 '중분류'와 '주요제품' 조합별로 데이터의 개수(빈도)를 구합니다.
combination_freq = df.groupby(['중분류', '주요제품']).size().reset_index(name='빈도')

# 2. 각 '중분류'의 전체 데이터 수(빈도)를 구합니다.
total_freq = df['중분류'].value_counts().reset_index()
total_freq.columns = ['중분류', '전체빈도']

# 조합별 빈도와 전체 빈도를 병합합니다.
merged_freq = pd.merge(combination_freq, total_freq, on='중분류')

# 3. 상대 빈도를 계산합니다.
merged_freq['산업내 주력상품 판매비중'] = (merged_freq['빈도'] / merged_freq['전체빈도']) * 100

# 상대 빈도의 합계를 확인
sum_percentage = merged_freq.groupby('중분류')['산업내 주력상품 판매비중'].sum()

if sum_percentage.min() < 99.9 or sum_percentage.max() > 100.1:
    print("중분류별 합계가 100%가 아닌 경우가 있습니다.")
    print(sum_percentage)
else:
    print("모든 중분류의 주요제품 판매비중 합계는 100%입니다.")
```

```python
# '중분류'와 '주요제품'을 기준으로 병합
df = pd.merge(df, merged_freq[['중분류', '주요제품', '산업내 주력상품 판매비중']], on=['중분류', '주요제품'], how='left')
df
```

# 1. 표준화: 최소-최대 스케일링(standardize)
### 장점: 모든 데이터를 지정된 범위(예: 1~10) 내로 변환
### 단점: 이상치에 민감, 이상치 때문에 대부분의 데이터가 좁은 범위 내로 몰릴 수 있음
### 적용 시 고려 사항: 그래프에서 이상치가 많이 관찰. 이 방법을 사용할 경우 다른 값들이 좁은 범위로 몰릴 가능성이 있음

# 2. 표준화: Robust Scaler 
### 장점: 중앙값과 IQR을 사용하여 스케일링하므로 이상치에 덜 민감
### 단점: 데이터의 최소 및 최대 값이 지정된 범위(예: 1~10)에 정확히 들어가지 않을 수 있음
### 적용 시 고려 사항: 그래프에서 이상치와 치우침이 관찰되는 경우 사용

```python
from sklearn.preprocessing import RobustScaler

# Robust Scaler 객체 생성
scaler = RobustScaler()

# '산업내 주력상품 판매비중' 컬럼만 스케일링
df_copy1['표준화'] = scaler.fit_transform(df_copy1[['산업내 주력상품 판매비중']])

# 스케일링된 값의 최소와 최대값을 찾아 1~10 범위로 조정
min_val = df_copy1['표준화'].min()
max_val = df_copy1['표준화'].max()

# 값을 1~10 범위로 조정
df_copy1['표준화'] = 1 + (df_copy1['표준화'] - min_val) * (10 - 1) / (max_val - min_val)

# 값을 반올림하고 정수로 변환
df_copy1['표준화'] = df_copy1['표준화'].round().astype(int)
df_copy1[['중분류', '주요제품', '산업내 주력상품 판매비중', '표준화']]
```
