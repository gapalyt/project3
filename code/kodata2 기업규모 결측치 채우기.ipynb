```python
# '기업규모' 열에서 공백 제거 및 콤마 뒤 글자 처리 함수 정의
def clean_company_size(value):
    if '폐업' in value:
        return value
    else:
        parts = value.split(',')
        return parts[0].strip() if len(parts) > 1 else value.strip()

# '기업규모' 열에 함수 적용
combined_df['기업규모'] = combined_df['기업규모'].apply(clean_company_size)
combined_df[['기업규모']]
```
