### 1.직접수출여부
- 수출여부(EU,US)사용해서 통합

```python
def export_status(row):
    if row['수출 여부 (EU)'] == '유' and row['수출 여부 (US)'] == '유':
        return 10
    elif row['수출 여부 (EU)'] == '유' or row['수출 여부 (US)'] == '유':
        return 5.5
    elif row['수출 여부 (EU)'] == '무' and row['수출 여부 (US)'] == '무':
        return 1
    else:
        return 0

# 새로운 '직접수출여부' 컬럼 생성
df['직접수출여부'] = df.apply(export_status, axis=1)
df.head(3)
```

### 4.최근 3년간 매출/영업이익
- (최근년도 매출액 - 3년전 매출액)/3년전 매출액, 매출액(영업이익)

```python
def calculate_growth(value_2019, value_2021):
    if value_2019 == 0:
        return 0  
    else:
        return (value_2021 - value_2019) / value_2019

df['매출추이'] = df.apply(lambda row: calculate_growth(row['매출액_2019'], row['매출액_2021']), axis=1)
df['영업이익 추이'] = df.apply(lambda row: calculate_growth(row['영업이익_2019'], row['영업이익_2021']), axis=1)
df.tail(3)
```

### 5.ESG_dept : ESG전담부서가 있을 가능성
- 메인비즈,이노비즈,벤처여부 3가지 이용하여 산정
- 다 없으면 1, 하나라도 있으면 4, 두개 있으면 7, 세개 다 있으면 10

```python
def calculate_esg_dept(row):
    yes_count = sum([row['메인비즈여부'] == '여', row['이노비즈여부'] == '여', row['벤처여부'] == '여'])
    
    if yes_count == 0:
        return 1
    elif yes_count == 1:
        return 4
    elif yes_count == 2:
        return 7
    elif yes_count == 3:
        return 10
    else :
         return 0

# 새로운 'ESG_dept' 컬럼 생성
df['ESG_dept'] = df.apply(calculate_esg_dept, axis=1)
df.head(3)
```
